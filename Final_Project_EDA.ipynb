# Compare of (Gender & Years in company) for all States


```python
# compare the males & females of company in all states for how long them stay in it.
sns.set(style="whitegrid")

xs = df['State']
hue = df['Gender']
ys = df['Age in Company (Years)']
plt.figure(figsize=(17, 8))
g = sns.barplot(x=xs, y=ys, hue=hue,ci=None )
g.set_yscale("linear")
_ = g.set(xlabel="States", ylabel="Number of Years")
```
